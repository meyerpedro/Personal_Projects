```python
import requests
import json
```

```python
response = requests.get("https://api.covid19api.com/summary").text
```

```python
response
```

```python
response_info = json.loads(response)
```

```python
response_info
```

```python
country_lst = []
for country_info in response_info['Countries']:
    country_lst.append([country_info['Country'],
                        country_info['NewDeaths'],
                        country_info['TotalDeaths'],
                       country_info['Date']])
```

```python
import pandas as pd
```

```python
covid_deaths = pd.DataFrame(data=country_lst, 
                            columns=['country', 'new_deaths', "total_deaths", 'date'])
```

```python
covid_deaths['date'] = pd.to_datetime(covid_deaths['date'])
```

```python
output_array = []

config = {
    'length':100,
}

config2 = {
    'length':200
}
config3 = {
    'length': 20,
    'number_columns': [
        {
            'name':'new_deaths',
            'min':0,
            'max':5000
        },
        {
            'name':'total_deaths',
            'min':0,
            'max':500000
        }]
}
```

```python
def single_random_obj(config, id):
    
    country = np.random.choice(covid_deaths['country'].tolist())
    number_columns = config['number_columns']
    obj = {
        'case_id':id,
        'country': country,
    }
        
    for i in number_columns:
        k = i['name']
        v = np.random.randint(i['min'], i['max'])
        obj[k]=v
        
    return obj
```

```python
for i in range(config3['length']):
    obj = single_random_obj(config3, i)
    output_array.append(obj)
```

```python
obj = pd.DataFrame(data=output_array, 
                   columns=['id', 'country', 'new_deaths', "total_deaths"] )
```
